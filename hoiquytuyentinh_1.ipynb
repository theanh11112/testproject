```python
import numpy as np
import matplotlib.pyplot as plt

# Dữ liệu
x = np.array([1, 2, 3, 4, 5, 6])  # Giờ học
y = np.array([50, 55, 60, 65, 70, 75])  # Điểm kiểm tra

# Bước 1: Tạo ma trận thiết kế X (thêm bias)
X = np.vstack([np.ones(len(x)), x]).T  # Ma trận X: [1, x]
print("Ma trận X (Thiết kế):\n", X)

# Tính hệ số hồi quy (theta) bằng công thức chuẩn tắc (pseudo-inverse)
print("ma tran X.T \n",X.T)
print('ma tran X.T @ X \n ',X.T @ X)
print("ma tran X.T @ X chuyen vi \n ",np.linalg.pinv(X.T @ X))
print("ma tran X.T @ y \n",X.T @ y)
theta = np.linalg.pinv(X.T @ X) @ X.T @ y  # Sử dụng pseudo-inverse
print("theta \n ",theta)
# Bước 2: Tính toán hệ số hồi quy (theta) bằng công thức chuẩn tắc
theta = np.linalg.inv(X.T @ X) @ X.T @ y
w0, w1 = theta  # w0 (intercept), w1 (slope)

print("Hệ số hồi quy:")
print(f"w0 (hệ số chặn): {w0}")
print(f"w1 (hệ số góc): {w1}")

# Bước 3: Dự đoán giá trị y_hat
y_pred = X @ theta  # y_hat = X @ theta

# Bước 4: Tính sai số MSE
mse = np.mean((y - y_pred) ** 2)
print(f"Sai số trung bình bình phương (MSE): {mse}")

# Bước 5: Trực quan hóa
plt.scatter(x, y, color='blue', label='Dữ liệu thực')  # Dữ liệu thực
plt.plot(x, y_pred, color='red', label='Hồi quy tuyến tính')  # Đường hồi quy
plt.xlabel('Giờ học (x)')
plt.ylabel('Điểm kiểm tra (y)')
plt.title('Hồi quy tuyến tính')
plt.legend()
plt.grid()
plt.show()
```
