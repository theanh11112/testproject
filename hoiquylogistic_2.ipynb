```python
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

# Dữ liệu
X = np.array([[1, 2], [2, 3], [3, 1], [4, 2], [5, 3]])  # 2 biến đầu vào
y = np.array([0, 0, 1, 1, 1])                           # Nhãn nhị phân

# Huấn luyện mô hình
model = LogisticRegression()
model.fit(X, y)

# Tham số mô hình
w0 = model.intercept_[0]
w1, w2 = model.coef_[0]
print(f"w0: {w0}, w1: {w1}, w2: {w2}")

# Trực quan hóa
x1_values = np.linspace(0, 6, 100)
x2_values = -(w1 / w2) * x1_values - (w0 / w2)

plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='red', label='Label 0')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='blue', label='Label 1')
plt.plot(x1_values, x2_values, color='green', label='Decision Boundary')

# Thiết lập đồ thị
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Logistic Regression Decision Boundary')
plt.legend()
plt.show()
```
