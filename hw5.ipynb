```python
# Import necessary libraries
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

# Data points and labels
X = np.array([[1, 2], [2, 3], [3, 1], [2, 5], [4, 3]])
y = np.array([1, 1, 1, 0, 0])

# Initialize and train the logistic regression model
model = LogisticRegression()
model.fit(X, y)

# Coefficients of the separating line
w1, w2 = model.coef_[0]
b = model.intercept_[0]

# Equation of the separating line: w1*x1 + w2*x2 + b = 0
# Rearrange to get x2 in terms of x1: x2 = -(w1/w2) * x1 - b/w2
x_values = np.linspace(0, 5, 100)
y_values = -(w1/w2) * x_values - b/w2

# Predict the label of the new point (1, 4)
point = np.array([[1, 4]])  # The point you want to predict
predicted_label = model.predict(point)

# Plot the data points and the separating line
plt.scatter(X[y==1][:, 0], X[y==1][:, 1], color='blue', marker = '*', label='Label 1')
plt.scatter(X[y==0][:, 0], X[y==0][:, 1], color='red', label='Label 0')
plt.plot(x_values, y_values, label='Separating line', color='green')

# Plot the new point (1,4) and annotate its predicted label
plt.scatter(point[0][0], point[0][1], color='blue', marker='x', s=100, label='Point (1,4)')
plt.text(1, 4.2, f'Label: {predicted_label[0]}', fontsize=12, color='purple')

# Set plot labels and title
plt.xlabel('Feature 1 (x1)')
plt.ylabel('Feature 2 (x2)')
plt.title('Logistic Regression: Predicting Label for Point (1,4)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
X = np.array([[1 , 2], [2 , 3] , [3 , 1] , [2 , 5] , [4 , 3]])
```

```python
Y = np.array([1 , 1 , 1 , 0 , 0])
```

```python
def loss(t):
    L = 0
    for x , y in zip(X , Y):
        z = t[0] + t[1] * x[0] + t[2] * x[1]
        if y == 1:
            L += np.log(sigmoid(z))
        else:
            L +=  np.log(1-sigmoid(z))
```

```python
def gradient(t):
    g = [0 , 0 , 0]
    for x , y in zip (X , Y):
        z = t[0] + t[1] * x[0] + t[2] * x[1]
        g[0] += sigmoid(z) * np.exp(-z)
        g[1] += sigmoid(z) * np.exp(-z) * x[0]
        g[2] += sigmoid(z) * np.exp(-z) * x[1]
    return g
```
