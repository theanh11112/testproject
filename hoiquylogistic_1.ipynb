```python
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

# Dữ liệu
X = np.array([[1], [2], [3], [4], [5]])  # Biến đầu vào 1 chiều
y = np.array([0, 0, 1, 1, 1])           # Nhãn nhị phân

# Huấn luyện mô hình
model = LogisticRegression()
model.fit(X, y)

# Tham số mô hình
w0 = model.intercept_[0]  # w0 là bias
w1 = model.coef_[0][0]    # w1 là hệ số cho x
print(f"w0: {w0}, w1: {w1}")

# Trực quan hóa
x_values = np.linspace(0, 6, 100)
y_values = 1 / (1 + np.exp(-(w0 + w1 * x_values)))  # Hàm sigmoid

plt.scatter(X, y, color='red', label='Data points')
plt.plot(x_values, y_values, label='Logistic curve', color='blue')

# Thiết lập đồ thị
plt.xlabel('x')
plt.ylabel('Probability')
plt.title('Logistic Regression: One Variable')
plt.legend()
plt.show()
```
