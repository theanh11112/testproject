```python
import numpy as np
import matplotlib.pyplot as plt

# Dữ liệu
x1 = np.array([1, 1, 2, 2, 3, 3])
x2 = np.array([1, 2, 1, 2, 1, 2])
y = np.array([6, 11, 8, 15, 12, 19])

# Tạo ma trận thiết kế X cho hồi quy đa thức bậc 2
X = np.vstack([
    np.ones(x1.shape[0]),  # Cột 1 (hằng số)
    x1,                    # Cột x1
    x2,                    # Cột x2
    x1**2,                 # Cột x1^2
    x1 * x2,               # Cột x1 * x2
    x2**2                  # Cột x2^2
]).T
print("Ma trận X (Thiết kế):\n", X)
print("ma tran X.T \n",X.T)
print('ma tran X.T @ X \n ',X.T @ X)
print("ma tran X.T @ X chuyen vi \n ",np.linalg.pinv(X.T @ X))
print("ma tran X.T @ y \n",X.T @ y)
# Tính toán các hệ số hồi quy theta
theta = np.linalg.inv(X.T @ X) @ X.T @ y
print("theta \n",theta)
# Dự đoán y
y_pred = X @ theta

# Tính sai số MSE
mse = np.mean((y_pred - y)**2)

# In kết quả
print(f"Tham số theta: {theta}")
print(f"MSE: {mse}")

# Trực quan hóa
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Vẽ dữ liệu thật
ax.scatter(x1, x2, y, color='blue', label='Dữ liệu thật')

# Vẽ dữ liệu dự đoán
ax.scatter(x1, x2, y_pred, color='red', label='Dữ liệu dự đoán')

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
plt.legend()
plt.show()
```
