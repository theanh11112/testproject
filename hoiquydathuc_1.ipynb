```python
import numpy as np
import matplotlib.pyplot as plt

# Dữ liệu
x = np.array([0, 1, 2, 3, 4])
y = np.array([1, 2, 6, 12, 20])


# Tạo ma trận thiết kế (X) cho hồi quy đa thức bậc 2
X = np.vstack([np.ones(x.shape[0]), x, x**2]).T
print("Ma trận X (Thiết kế):\n", X)

print("ma tran X.T \n",X.T)
print('ma tran X.T @ X \n ',X.T @ X)
print("ma tran X.T @ X chuyen vi \n ",np.linalg.pinv(X.T @ X))
print("ma tran X.T @ y \n",X.T @ y)

# Tính toán các hệ số hồi quy theta
theta = np.linalg.inv(X.T @ X) @ X.T @ y
print("theta \n",theta)

# Dự đoán y
y_pred = X @ theta

# In kết quả
print(f"Tham số theta: {theta}")

# Vẽ đồ thị
plt.scatter(x, y, color='blue', label='Dữ liệu thật')
plt.plot(x, y_pred, color='red', label='Dự đoán hồi quy đa thức')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
